```python
import pandas as pd
import plotly.express as px

# Read the CSV file
data = pd.read_csv('revised.csv')
data = data[data.country == "GT"]
# Group the data by 'avg_log_income' and aggregate the sum of 'migrated_yn_1' and 'total_count'
grouped_data = data.groupby('avg_log_income').agg({'migrated_yn_1': 'sum', 'total_count': 'sum'}).reset_index()

# Calculate the ratio of the summed 'migrated_yn_1' to the summed 'total_count'
grouped_data['migrated_yn_1_ratio'] = grouped_data['migrated_yn_1'] / grouped_data['total_count']

# Create a bar plot
fig = px.bar(grouped_data,
             x='avg_log_income',
             y='migrated_yn_1_ratio',
             labels={
                 'avg_log_income': 'Average Log Income',
                 'migrated_yn_1_ratio': 'Migrated Ratio (Sum of migrated_yn_1 / Sum of total_count)'
             },
             title='Bar Plot: Migrated Ratio vs Average Log Income')

# Update the layout to set the background color to white
fig.update_layout(
    plot_bgcolor='rgba(255, 255, 255, 1)',
    paper_bgcolor='rgba(255, 255, 255, 1)',
)

# Show the plot
fig.show()
fig.write_html('1-GT.html')
```

```python
import pandas as pd
import plotly.graph_objs as go
from plotly.subplots import make_subplots

# Read the CSV file
data = pd.read_csv('sample2.csv')

# Define the column_list (replace this with your actual list of columns)
column_list = ['log_income_per_capita', 'income_satisfaction', 'income_sufficiency_6m', 'econ_condition']

# Choose one of the columns from column_list (replace 'column1' with your desired column)
selected_column = 'column1'

# Filter the data for migrated_yn = 0 and migrated_yn = 1
data_migrated_0 = data[data['migrated_yn'] == 0]
data_migrated_1 = data[data['migrated_yn'] == 1]

# Create a subplot with 1 row and 2 columns
fig = make_subplots(rows=1, cols=2, subplot_titles=("Migrated_yn = 0", "Migrated_yn = 1"))

# Add histograms to the subplots
fig.add_trace(go.Histogram(x=data_migrated_0['avg_log_income'],
                           y=data_migrated_0[selected_column],
                           histfunc="sum",
                           name="Migrated_yn = 0"),
              row=1, col=1)

fig.add_trace(go.Histogram(x=data_migrated_1['avg_log_income'],
                           y=data_migrated_1[selected_column],
                           histfunc="sum",
                           name="Migrated_yn = 1"),
              row=1, col=2)

# Update the layout
fig.update_layout(
    title="Histogram: Total Count vs Average Log Income",
    xaxis_title="Average Log Income",
    yaxis_title="Total Count",
    plot_bgcolor='rgba(255, 255, 255, 1)',
    paper_bgcolor='rgba(255, 255, 255, 1)',
    xaxis=dict(showgrid=True, gridcolor='black', linewidth=1, linecolor='black'),
    yaxis=dict(showgrid=True, gridcolor='black', linewidth=1, linecolor='black')
)

# Show the plot
fig.show()
```

```python
import plotly.subplots as sp
import plotly.graph_objs as go
import pandas as pd

# Replace this with your actual DataFrame
df = pd.read_csv("sample2.csv")
df = df.drop(df[df.eq(99)].dropna(how='all').index)
df = df[df.country == "SLV"]

# Replace this with your actual list of columns
column_list = ['avg_log_income','log_income_per_capita', 'income_satisfaction', 'income_sufficiency_6m', 'econ_condition']

# Filter the data based on migrated_yn
migrated_0 = df[df["migrated_yn"] == 0]
migrated_1 = df[df["migrated_yn"] == 1]

# Set the number of bins
num_bins = 20  # Change this value to set the desired number of bins

# Create subplots with 1 row and 2 columns
fig = sp.make_subplots(rows=1, cols=2, subplot_titles=("migrated_yn = 0", "migrated_yn = 1"))

colors = ["blue", "red", "green", "black", "purple"]  # You can customize these colors

# Loop through the columns and create traces for each column
for i, column in enumerate(column_list):
    # Calculate the bin size
    bin_size = (migrated_0[column].max() - migrated_0[column].min()) / num_bins

    trace1 = go.Histogram(
        x=migrated_0[column],
        histnorm='',
        name=column,
        legendgroup=column,
        marker=dict(color=colors[i]),
        opacity=0.75,
        visible=True if i == 0 else "legendonly",
        xbins=dict(start=migrated_0[column].min(), end=migrated_0[column].max(), size=bin_size),
    )

    # Calculate the bin size
    bin_size = (migrated_1[column].max() - migrated_1[column].min()) / num_bins

    trace2 = go.Histogram(
        x=migrated_1[column],
        histnorm='',
        name=column,
        showlegend=False,
        legendgroup=column,
        marker=dict(color=colors[i]),
        opacity=0.75,
        visible=True if i == 0 else "legendonly",
        xbins=dict(start=migrated_1[column].min(), end=migrated_1[column].max(), size=bin_size),
    )
    fig.add_trace(trace1, row=1, col=1)
    fig.add_trace(trace2, row=1, col=2)

# Update layout
fig.update_layout(
    title_text="Histograms by migrated_yn",
    xaxis_title_text="Column",
    yaxis_title_text="Count",
    bargap=0.2,
    bargroupgap=0.1
)

# Show the plot
fig.show()
fig.write_html("histograms_yes_no_SLV.html")
```

```python
import pandas as pd
import plotly.graph_objs as go
import plotly.subplots as sp

# Replace this with your actual DataFrame
df = pd.read_csv("sample4.csv")

# Replace this with your actual list of columns
column_list = ["avg_log_income", "income_per_capita",	"mig_ext_cost_total_usd",
               "debt_amount",	"sucess_yn", "institutional_loan_yn"]


# Create a subplot with 1 row and 1 column
fig = sp.make_subplots(rows=1, cols=1)

# Define the x-axis as avg_log_income
x_axis_column = "mig_ext_attempts"

# Create a trace for each column
for i, column in enumerate(column_list):
    trace = go.Bar(
        x=df[x_axis_column],
        y=df[column],
        name=column,
        visible=True if i == 0 else "legendonly",
    )

    fig.add_trace(trace)

# Update layout
fig.update_layout(
    title_text="Bar chart with toggleable y-axis",
    xaxis_title_text=x_axis_column,
    yaxis_title_text="Value",
    barmode="group",
    updatemenus=[dict(
        buttons=list([
            dict(
                label=column,
                method="update",
                args=[{"visible": [True if j == i else False for j in range(len(column_list))]}, {"title": f"Bar chart with toggleable y-axis - {column}"}]
            ) for i, column in enumerate(column_list)
        ]),
        direction="down",
        showactive=True,
        xanchor="left",
        yanchor="top",
        x=0.01,
        y=1.05,
    )],
)

# Show the plot
fig.show()
fig.write_html("attempts_charts.html")

```
